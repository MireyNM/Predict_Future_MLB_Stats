# Import libraries

```python
# Load necessary libraries
import pandas as pd 
```

# Load raw Data for Pitchers from 2017 till 2022

```python
# Load data from CSV files
pitch_2017 = pd.read_csv("../Data/Raw_Data/pitch_perc2017.csv")
pitch_2018 = pd.read_csv("../Data/Raw_Data/pitch_perc2018.csv")
pitch_2019 = pd.read_csv("../Data/Raw_Data/pitch_perc2019.csv")
pitch_2020 = pd.read_csv("../Data/Raw_Data/pitch_perc2020.csv")
pitch_2021 = pd.read_csv("../Data/Raw_Data/pitch_perc2021.csv")
pitch_2022 = pd.read_csv("../Data/Raw_Data/pitch_perc2022.csv")
```

```python
# Adding a year column to each DataFrame
pitch_2017.insert(0,'Year',2017)
pitch_2018.insert(0,'Year',2018)
pitch_2019.insert(0,'Year',2019)
pitch_2020.insert(0,'Year',2020)
pitch_2021.insert(0,'Year',2021)
pitch_2022.insert(0,'Year',2022)
```

# Preparing the Data

```python
# Concatinate 6 DataFrames in one 
pitch_all=pd.concat([pitch_2017, pitch_2018, pitch_2019, pitch_2020, pitch_2021, pitch_2022], axis=0)
pitch_all
```

```python
# Rename Columns 
renamed_pitch = pitch_all.rename(columns={
    "Year":"Year",
    "last_name":"Last_Name",
    "first_name": "First_Name",
    "pitcher":"Pitcher_ID",
    "n_ff":"4_Seamer",
    "n_si":"Sinker",
    "n_fc":"Cutter",
    "n_sl":"Slider",
    "n_ch": "Changeup",
    "n_cu": "Curve",
    "n_fs": "Splitter",
    "n_kn":"Knuckle" })
renamed_pitch.head()
```

```python
# Drop pitcher ID column 
pitch_df=renamed_pitch.drop(['Pitcher_ID'], axis = 1, inplace = False)
pitch_df
```

To make means simpler, we are going to combine the 4-seam and cutter as **fast-pitch**, and the rest of pitchs' types as **slow-pitch**.

```python
# Grouping 4-Seam and cutter as "Fast-pitch"
pitch_df['Fast-pitch']= pitch_df['4_Seamer'] + pitch_df['Cutter']
pitch_df
```

```python
# Grouping all other pitches' types as "Slow-pitch"
pitch_df['Slow-pitch']= pitch_df['Sinker'] + pitch_df['Slider'] + pitch_df['Changeup'] + pitch_df['Curve'] + pitch_df['Splitter'] + pitch_df['Knuckle']
pitch_df.head()
```

```python
# Drop pitcher ID column 
pitch_perc_df = pitch_df.drop(['4_Seamer','Sinker','Cutter','Slider','Changeup','Curve','Splitter','Knuckle'], axis = 1, inplace = False)
pitch_perc_df
```

```python
# Export DataFrame to csv
pitch_perc_df.to_csv('../Data/Clean_Data/pitch_perc.csv', index=False)
```
