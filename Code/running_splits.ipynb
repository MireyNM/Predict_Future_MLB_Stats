```python
# Importing dependiences 
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

```python
# Importing data from csv files by year
data_2022 = pd.read_csv("../Data/Raw_Data/running_splits_2022.csv")
data_2021 = pd.read_csv("../Data/Raw_Data/running_splits_2021.csv")
data_2020 = pd.read_csv("../Data/Raw_Data/running_splits_2022.csv")
data_2019 = pd.read_csv("../Data/Raw_Data/running_splits_2022.csv")
data_2018 = pd.read_csv("../Data/Raw_Data/running_splits_2018.csv")
data_2017 = pd.read_csv("../Data/Raw_Data/running_splits_2017.csv")
data_2016 = pd.read_csv("../Data/Raw_Data/running_splits_2016.csv")
data_2015 = pd.read_csv("../Data/Raw_Data/running_splits_2015.csv")
```

```python
# Inserting new column 'year' in each table
data_2022.insert(0, 'year', 2022)
data_2021.insert(0, 'year', 2021)
data_2020.insert(0, 'year', 2020)
data_2019.insert(0, 'year', 2019)
data_2018.insert(0, 'year', 2018)
data_2017.insert(0, 'year', 2017)
data_2016.insert(0, 'year', 2016)
data_2015.insert(0, 'year', 2015)
```

```python
# Joining tables into the running_splits_dataframe
running_splits_df = pd.concat([data_2022, 
                              data_2021, 
                              data_2020,
                              data_2019, 
                              data_2018,
                              data_2017,
                              data_2016,
                              data_2015])
running_splits_df
```

```python
# Dropping unnessasry columns
running_splits_df_clean = running_splits_df.drop(['player_id',
                       'team_id',
                       'name_abbrev',
                       'bat_side',
                       'position_name',
                       'age'], axis = 1)
running_splits_df_clean
                       
```
