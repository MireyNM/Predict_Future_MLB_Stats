```python
# Importing dependiences 
import pandas as pd
import numpy as np
```

```python
# Importing data from csv files by year
data_2022 = pd.read_csv("../static/data/Raw_Data/running_splits_2022.csv")
data_2021 = pd.read_csv("../static/data/Raw_Data/running_splits_2021.csv")
data_2020 = pd.read_csv("../static/data/Raw_Data/running_splits_2022.csv")
data_2019 = pd.read_csv("../static/data/Raw_Data/running_splits_2022.csv")
data_2018 = pd.read_csv("../static/data/Raw_Data/running_splits_2018.csv")
data_2017 = pd.read_csv("../static/data/Raw_Data/running_splits_2017.csv")
data_2016 = pd.read_csv("../static/data/Raw_Data/running_splits_2016.csv")
data_2015 = pd.read_csv("../static/data/Raw_Data/running_splits_2015.csv")
```

```python
# Inserting new column 'year' in each table
data_2022.insert(0, 'year', 2022)
data_2021.insert(0, 'year', 2021)
data_2020.insert(0, 'year', 2020)
data_2019.insert(0, 'year', 2019)
data_2018.insert(0, 'year', 2018)
data_2017.insert(0, 'year', 2017)
data_2016.insert(0, 'year', 2016)
data_2015.insert(0, 'year', 2015)
```

```python
# Joining tables into the running_splits_dataframe
running_splits_df = pd.concat([data_2022, 
                              data_2021, 
                              data_2020,
                              data_2019, 
                              data_2018,
                              data_2017,
                              data_2016,
                              data_2015])
running_splits_df
```

```python
# Adding new column
running_splits_df.insert(28, 'predicted_time_outcome', "")
running_splits_df
```

```python
# Creating a new df_clean to hold data
running_splits_new = pd.DataFrame(running_splits_df)
```

```python
# Dropping columns
running_splits_df_clean = running_splits_new.drop(['player_id',
                            'name_abbrev',
                            'team_id',
                            'position_name',
                            'age',
                            'bat_side'], axis = 1)
running_splits_df_clean
```

```python
# meters per second for players
d = 90 #ft
t = running_splits_df['seconds_since_hit_090']
s = d/t

# Creating new column to table and filling with values
running_splits_df_clean.insert(22,'player_speed_(f/s)', s)
running_splits_df_clean
```

```python
# Checking df
running_splits_df_clean.head()
```

```python
# Exporting cleaned data
running_splits_df_clean.to_csv("../Data/Clean_Data/running_splits_clean.csv", index = False)
```

```python
# Initializing model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# Creating X, y components
X = running_splits_df_clean.copy()
X_dropped = X.drop(columns = ['predicted_time_outcome',
                              'year',
                             'last_name',
                             ' first_name',
                             'seconds_since_hit_090',
                             'player_speed_(f/s)'], axis = 1)

# Holding 'actual' results in y variable
y = pd.DataFrame(X['seconds_since_hit_090'])
y.head()
```

```python
# Scalling the dat
#from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
X_scaled = StandardScaler().fit_transform(X_dropped)
X_scaled = pd.DataFrame(X_scaled)
```

```python
# spliting the df into test and train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state = 1)

# Printing shape of test series
print(y_train.shape)
print(X_train.shape)
```

```python
# Fiting model to split data
fit_model = model.fit(X_train, y_train)
print(fit_model)
```

```python
# Using fited model to create predicted results of the runners time
y_pred_train = fit_model.predict(X_train)
y_pred_test = fit_model.predict(X_test)
print(y_pred_train)
print(y_pred_test)
```

```python
# Importing mse and r2 metrics
from sklearn.metrics import mean_squared_error, r2_score

# Creating variables to hold mse and r2 scores
mse = mean_squared_error(y_test, y_pred_test)
r2 = r2_score(y_test, y_pred_test)
print(mse)
print(r2)
```

```python
# Filling empty DataFrame with actual/predicted values
# Creating DataFrame
df = pd.DataFrame()

# Filling columns
df['actual'] = y_test['seconds_since_hit_090']
df['prediction'] = y_pred_test
df.reset_index(inplace = False)
df.head(10)
```
